3.Create the Agent Class

```python
class ConciergeAgent:
    def __init__(self, model_name="gemini-1.5-pro"):
        self.model = genai.GenerativeModel(model_name)

    def generate(self, prompt):
        response = self.model.generate_content(prompt)
        return response.text

    def summarize_text(self, text):
        prompt = f"Summarize the following text in simple, clear bullet points:\n\n{text}"
        return self.generate(prompt)

    def draft_email(self, subject, details):
        prompt = f"""
        Write a professional email.
        Subject: {subject}
        Details: {details}

        Keep tone polite, clear, respectful.
        """
        return self.generate(prompt)

    def prioritize_tasks(self, tasks):
        prompt = f"""
        You are a productivity agent. Prioritize these tasks from highest-impact to lowest-impact:

        {tasks}

        Output format:
        1. Task — Priority reason
        2. Task — Priority reason
        """
        return self.generate(prompt)

    def decision_helper(self, question, options):
        prompt = f"""
        Help me make a decision.

        Question: {question}
        Options: {options}

        Provide:
        - Pros/cons for each option
        - Best recommendation
        """
        return self.generate(prompt)
```

4.Instantiate the Agent

```python
agent = ConciergeAgent()
print("AI Concierge Agent is ready!")
```

5.Use Case 1: Smart Task Prioritization




```python
tasks = """
- Finish project report
- Clean my room
- Prepare for client meeting
- Learn SQL
- Pay electricity bill
"""

print(agent.prioritize_tasks(tasks))
```
